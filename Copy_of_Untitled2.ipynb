```
import matplotlib.pyplot as plt
import numpy as np 
from numpy import random
from scipy.fft import fft, ifft 
import math


min = -50                                # Starting point 
max = 50                                 # Ending point
step= 0.001                              # The resolution
A   = 22                                 #Amplitude factor
p = int(abs((max-min)/step))
x= np.arange(min,max,step)
lamda = 0.9*np.pi                       # Wavelength of E inside the envelope 
k = 2*np.pi/lamda
sin = np.sin(k*x)

sigma = 9                               # spread - standard deviation
m     = 3                               # location of the peak - mean 
expo  = -0.5*((x-m)*(x-m)*(1/sigma)*(1/sigma))
nfactor = 0.1                         # Random noise in the output
noise1 = np.random.rand(p)*nfactor
noise2 = np.random.rand(p)*nfactor

gaus = (1/sigma*math.sqrt(2*np.pi))*np.exp(expo)
out = (A*gaus*sin) + noise1 - noise2
plt.plot(out)
plt.show()
plt.plot(np.abs(fft(out)))
```
